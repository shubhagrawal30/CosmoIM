```python
import pandas as pd
df = pd.read_csv('./CosmoAstroSeed_IllustrisTNG.txt', sep='\s+')
df
```

```python
sequence = []
for i in range(1, 7):
    for j in range(-5, 6):
        n = f'n{abs(j)}' if j < 0 else f'{j}'
        sequence += [f'1P_{i}_{n}']
        
sequence
```

```python
df.loc[df["Name"].str.startswith("1P_"), "Name"] = sequence
```

```python
df[df["Name"].str.startswith("1P_")]
```

```python
df.to_csv('camels_parameters.csv', index=False, float_format='%.5f')
```
