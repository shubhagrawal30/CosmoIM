```python
import os
import shutil

source_dir = "/global/cfs/cdirs/des/shubh/timsim/simim_resources/simulations/IllustrisTNG"
dest_dir = "/global/cfs/cdirs/des/shubh/timsim/simim_resources/simulations/camels"

for root, dirs, files in os.walk(source_dir):
    print(root[root.rindex('/'):], end="\t")
    for filename in files:
        if filename.startswith("fof_subhalo_tab_") and filename.endswith(".hdf5"):
            file_path = os.path.join(root, filename)
            dest_path = file_path.replace(source_dir, dest_dir).replace("IllustrisTNG", "").replace("fof_subhalo_tab_", "/raw/groups_")
            dest_path = os.path.join(dest_path[:-5], "fof_subhalo_tab_" + dest_path[-8:])
            dest_dirname = os.path.dirname(dest_path)
            os.makedirs(dest_dirname, exist_ok=True)
            shutil.copyfile(file_path, dest_path)
```
