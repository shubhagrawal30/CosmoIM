```python
# set the source and destination directories
src_dir = "/global/cfs/cdirs/des/shubh/timsim/simim_resources/simulations/camels_CV_0/"
dest_dir = "/global/cfs/cdirs/des/shubh/timsim/simim_resources/simulations/camels_CV_0/raw/groups_"

# iterate over all the files in the source directory
for filename in os.listdir(src_dir):
    
    # check if the file is of the desired format
    if filename.startswith("fof_subhalo_tab_") and filename.endswith(".hdf5"):
        
        # extract the three digit number from the filename
        num = filename[17:20]
        
        # create the destination directory if it doesn't exist
        dest_subdir = dest_dir + num
        os.makedirs(dest_subdir, exist_ok=True)
        
        # move the file to the destination directory
        src_file = os.path.join(src_dir, filename)
        dest_file = os.path.join(dest_subdir, filename)
        shutil.move(src_file, dest_file)
```
